```python
import pandas as pd
import glob
import os

months = ['08. Aug/', '09. Sep/', '10. Oct/', '11. Nov/', '12. Dec/']
result = pd.DataFrame()
result_columns = ['Month', 'Garments prod', 'SMV', 'Efficiency', 'LIN prod', 'LIN SMV', 'LIN Efficiency']
result = result.reindex(columns=result_columns)
i = 0
for folder in months:
    path = 'Archive/2024/' + folder
    files = glob.glob(os.path.join(path, '*.xlsx'))
    print(files)
    latest_file = max(files, key=os.path.getctime)
    df = pd.read_excel(latest_file)
    for index, row in df.iterrows():
        if row[0] == 'GMT TOTAL:':
            result.loc[i, 'Month'] = folder[3:-1]
            result.loc[i, 'Garments prod'] = row[3]
            result.loc[i, 'SMV'] = row[14]
            result.loc[i, 'Efficiency'] = row[18]
        if row[0] == 'LINGERIE':
            result.loc[i, 'LIN prod'] = row[3]
            result.loc[i, 'LIN SMV'] = row[14]
            result.loc[i, 'LIN Efficiency'] = row[18]
    i += 1
result.to_excel('summary_report.xlsx', index=False)
```
